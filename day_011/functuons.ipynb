```python
import statistics
# 1. Declare a function named evens_and_odds . It takes a positive integer as parameter and it counts number of evens and odds in the number.
def count_even_odd(limit):
    even_count = 0
    odd_count = 0

    for num in range(1, limit + 1):
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count

# Count even and odd numbers up to 100
even, odd = count_even_odd(10)

print(f"The number of even numbers up to 100 is: {even}")
print(f"The number of odd numbers up to 100 is: {odd}")


# 2. Call your function factorial, it takes a whole number as a parameter and it return a factorial of the number
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# 3. Call your function is_empty, it takes a parameter and it checks if it is empty or not
def is_empty(param):
    return not bool(param)

# 4. Write different functions which take lists. They should calculate_mean, calculate_median, calculate_mode, calculate_range, calculate_variance, calculate_std (standard deviation).
def calculate_mean(lst):
    return statistics.mean(lst)

def calculate_median(lst):
    return statistics.median(lst)

def calculate_mode(lst):
    return statistics.mode(lst)

def calculate_range(lst):
    return max(lst) - min(lst)

def calculate_variance(lst):
    return statistics.variance(lst)

def calculate_std(lst):
    return statistics.stdev(lst)

print('2.', factorial(5)) 
print('3.', is_empty([]))  
print('4:')
numbers = [2, 4, 4, 4, 5, 5, 7, 9]
print('Mean:', calculate_mean(numbers)) 
print('Medium:', calculate_median(numbers)) 
print('Mode:', calculate_mode(numbers)) 
print('Range:', calculate_range(numbers)) 
print('Variance:', calculate_variance(numbers))  
print('std:', calculate_std(numbers)) 
```

```python
import re
# 1. Write a function called is_prime, which checks if a number is prime.
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
print('1.', is_prime(17))
print('1.', is_prime(15))

# 2. Write a functions which checks if all items are unique in the list.
def are_all_unique(lst):
    return len(lst) == len(set(lst))
numbers = [1, 2, 3, 4, 5]
print('2.', are_all_unique(numbers))

# 3. Write a function which checks if all the items of the list are of the same data type.
def are_all_same_type(lst):
    return all(isinstance(elem, type(lst[0])) for elem in lst)

data_types = [1, 2, 3, 'four', 'five']
print('3.', are_all_same_type(data_types)) 

# 4. Write a function which check if provided variable is a valid python variable
def is_valid_variable(variable):
    return re.match(r'^[a-zA-Z_]\w*$', variable) is not None

variable_name = 'valid_variable'
print('4.', is_valid_variable(variable_name))

invalid_variable = '123invalid'
print('4.', is_valid_variable(invalid_variable))
```

```python
# 5. Go to the data folder and access the countries-data.py file.
# -Create a function called the most_spoken_languages in the world. It should return 10 or 20 most spoken languages in the world in descending order
# -Create a function called the most_populated_countries. It should return 10 or 20 most populated countries in descending order.

import countries_data
countries_data = countries_data.countries_data
# print(countries_data[45])

# Function to get the most spoken languages in the world
def most_spoken_languages(limit=10):
    language_count = {}
    
    for country in countries_data:
        languages = country['languages']
        for language in languages:
            if language in language_count:
                language_count[language] += 1
            else:
                language_count[language] = 1
    
    # Sort the languages by count in descending order
    sorted_languages = sorted(language_count.items(), key=lambda x: x[1], reverse=True)
    
    # Return the top 'limit' most spoken languages
    return sorted_languages[:limit]

# Function to get the most populated countries
def most_populated_countries(limit=10):
    countries_population = {}
    
    for country in countries_data:
        name = country['name']
        population = country['population']
        countries_population[name] = population
    
    # Sort the countries by population in descending order
    sorted_countries = sorted(countries_population.items(), key=lambda x: x[1], reverse=True)
    
    # Return the top 'limit' most populated countries
    return sorted_countries[:limit]

# Example usage:
print('20 Nost spoken languages:', most_spoken_languages(20))  # Get the top 10 most spoken languages
print('10 Most populated countries:', most_populated_countries(10))  # Get the top 10 most populated countries
```
