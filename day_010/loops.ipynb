```python
#1. Using a for loop to iterate from 0 to 100 and calculate the sum of all numbers
total_sum = 0
for i in range(101):
    total_sum += i

print(f"The sum of all numbers from 0 to 100 is: {total_sum}")

#2. Using a for loop to iterate from 0 to 100 and calculate the sum of even and odd numbers separately
sum_even = 0
sum_odd = 0

for i in range(101):
    if i % 2 == 0:  # Check if the number is even
        sum_even += i
    else:
        sum_odd += i

print(f"The sum of all even numbers is: {sum_even} and The sum of all odd numbers is: {sum_odd}")
print(f"")
```

```python
# 2. This is a fruit list, ['banana', 'orange', 'mango', 'lemon'] reverse the order using loop.
fruits = ['banana', 'orange', 'mango', 'lemon']
reversed_fruits = []

for fruit in reversed(fruits):
    reversed_fruits.append(fruit)

print("Reversed order of fruits:", reversed_fruits)
```
